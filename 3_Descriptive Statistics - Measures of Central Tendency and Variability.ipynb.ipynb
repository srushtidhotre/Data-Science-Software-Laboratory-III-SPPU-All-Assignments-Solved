```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic = titanic.dropna(subset=['age'])
```

```python
grouped_stats = titanic.groupby('sex')['age'].agg(['mean', 'median', 'min', 'max', 'std'])
print("Summary statistics of Age grouped by Sex:\n")
grouped_stats
```

```python
age_list_by_sex = titanic.groupby('sex')['age'].apply(list)
print("List of Ages for each Sex:\n")
age_list_by_sex
```

```python
iris = pd.read_csv('Iris.csv')
print("Iris dataset preview:\n")
iris.head()
```

```python
iris.columns
```

```python
species_stats = iris.groupby('Species').describe(percentiles=[.25, .5, .75])
print("Summary statistics for each species:\n")
species_stats
```

```python
# Group by 'Species' and calculate mean
mean_stats = iris.groupby('Species').mean()
print("Mean of each feature grouped by Species:\n")
mean_stats
```

```python
# Group by 'Species' and calculate median
median_stats = iris.groupby('Species').median()
print("Median of each feature grouped by Species:\n")
median_stats
```

```python
# Group by 'Species' and calculate standard deviation
std_stats = iris.groupby('Species').std()
print("Standard Deviation of each feature grouped by Species:\n")
std_stats
```

```python
# Group by 'Species' and calculate 25th percentile
q25_stats = iris.groupby('Species').quantile(0.25)
print("25th Percentile of each feature grouped by Species:\n")
q25_stats
```

```python
# Group by 'Species' and calculate 50th percentile (same as median)
q50_stats = iris.groupby('Species').quantile(0.50)
print("50th Percentile of each feature grouped by Species:\n")
q50_stats
```

```python
# Group by 'Species' and calculate 75th percentile
q75_stats = iris.groupby('Species').quantile(0.75)
print("75th Percentile of each feature grouped by Species:\n")
q75_stats
```
