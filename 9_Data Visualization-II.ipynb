```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

```python
# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic_data)
plt.title('Box plot of Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Plotting box plot for Age distribution by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='age', data=titanic_data)
plt.title('Box plot of Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Plotting box plot for Age distribution by Survival Status
plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='age', data=titanic_data)
plt.title('Box plot of Age Distribution by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.show()
```
